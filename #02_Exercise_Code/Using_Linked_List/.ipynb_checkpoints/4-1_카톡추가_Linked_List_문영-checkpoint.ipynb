```python
class Node:
    def __init__(self):
        self.data = None
        self.link = None
```

```python
def printNodes(start):
    current = start
    if current == None:
        return
    print(current.data, end=' ')
    while current.link != None:
        current = current.link
        print(current.data, end=' ')
    print()
```

```python
# ê°œì¡° ë°©ì‹ 
def insertNode(insertData):
    global memory, head, current, pre
    
    if insertData[1] >= head.data[1]:
        node = Node()
        node.data = insertData
        node.link = head
        head = node
        return
    
    current = head
    while current.link != None:
        pre = current
        current = current.link
        if insertData[1] >= current.data[1]:
            node = Node()
            node.data = insertData
            node.link = current
            pre.link = node
            return
    
    node = Node()
    node.data = insertData
    current.link = node
```

```python
def deleteNode(deleteData):
    global memory, head, current, pre
    
    if head.data == deleteData:
        current = head
        head = head.link
        del(current)
        return
    
    current = head
    while current.link != None:
        pre = current
        current = current.link
        if current.data == deleteData:
            pre.link = current.link
            del(current)
            return
```

```python
def findNode(findData):
    global memory, head, current, pre
    
    current = head
    if current.data == findData:
        return current
    while current.link != None:
        current = current.link
        if current.data == findData:
            return current
    return Node()
```
